```python
# Calculate the IQR for MSRP
Q1 = np.percentile(df.price, 25)
Q3 = np.percentile(df.price, 75)
IQR = Q3 - Q1

# Set a threshold for identifying outliers (e.g., values outside [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR])
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df.price < lower_bound) | (df.price > upper_bound)]

outliers[['floors','price']].head()
```

```python
# Cap Extreme values identified from Outliers
def cap_outliers(data, lower_threshold, upper_threshold):
    capped_data = np.clip(data, lower_threshold, upper_threshold)
    return capped_data

# Cap the outliers
capped_data = cap_outliers(df.price, lower_bound, upper_bound)
capped_data

df.loc[:,'price'] = capped_data
df.head()
```

```python
df1 = df[['price']]
df1

# Create subplots for the histograms
plt.figure(figsize=(12, 18))

for i, var in enumerate(df1, 1):
    plt.subplot(5, 2, i)
    sns.boxplot(x='floors', y=var, data=df)
    plt.title(f'Box Plot of {var}')
    plt.xlabel('floors')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
```

```python
# Select only categorical variables for the count plot
categorical_vars = ['condition','floors','view']

# Create subplots for the count plots
plt.figure(figsize=(12, 9))

for i, var in enumerate(categorical_vars, 1):
    plt.subplot(3, 2, i)
    sns.countplot(data=df, x=var)
    plt.title(f'Count Plot of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
x = df[['sqft_basement','bedrooms','bathrooms','floors','view']]
x
```

```python
y = df[['price']]
y
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
```

```python
# Create a linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train, y_train)

# Make predictions on the test data
y_pred = lr.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Coefficient of Determination): {round(r2,2)}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a linear regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to the training data
rf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Coefficient of Determination): {round(r2,2)}")
```

```python
 # save the model
import pickle
pickle.dump(rf, open("HousePrice_rf.pkl", "wb"))
```

```python
# load the model
rf = pickle.load(open("HousePrice_rf.pkl", "rb"))

# Make predictions on the New data
sqft_basement=25
bedrooms=2
bathrooms=2
floors=2
view=1


y_pred = rf.predict([[sqft_basement,bedrooms,bathrooms,floors,view]])

print(f"Predicted MSRP: {round(y_pred[0])}") 
```
